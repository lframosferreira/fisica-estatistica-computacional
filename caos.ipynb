```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1

N = 50

def plot(r: float, y0: float, N: int):
    y_values: np.array = np.zeros(N, dtype = float)
    y_values[0] = y0
    for n in np.arange(N - 1):
        y_values[n + 1] = r * y_values[n]* (1 - y_values[n])

    plt.plot(np.arange(N), y_values, '-', label = f"{y0}")
    plt.title(f"r = {r}")
    plt.xlabel("Instante n")
    plt.ylabel("População no instante n")

r: float = .5
y0: float = .5
plot(r, y0)
```

```python
# 2
y0: float = .5
for r in [2.5, 3.1, 3.5, 3.7]:
    plot(r, y0, N)
    plt.show()
```

```python
# 3
for r in [2.5, 3.1, 3.5, 3.7]:
    for y0 in [.25, .5, .75]:
        plot(r, y0, N)
    plt.legend()
    plt.show()
```

```python
# 4
for r in [2.5, 3.1, 3.5, 3.7]:
    for y0 in [.5, .501, .50001]:
        plot(r, y0, N)
    plt.legend()
    plt.show()
```

```python
# 5

N: int = 10000
r_values: np.array = np.linspace(10e-5, 4, num = N)
y0: float = .5
def plot_2(r_values: np.array, y0: float, N: int):
    for r in r_values:
        y_values: np.array = np.zeros(N, dtype = float)
        y_values[0] = y0
        for n in np.arange(N - 1):
            y_values[n + 1] = r * y_values[n]* (1 - y_values[n])
        plt.title("Diagrama de bifurcação")
        plt.plot([r] * 100, y_values[n - 98:], color = 'black', ls = '', marker = ',')
    plt.show()
plot_2(r_values, y0, N)
```

```python
# 6.1

r_values: np.array = np.linspace(3.7, 3.9, num = N)
y0: float = .5
plot_2(r_values, y0, N)
```

```python
# 6.2

r_values: np.array = np.linspace(3.84, 3.856, num = N)
y0: float = .5
plot_2(r_values, y0, N)
```
